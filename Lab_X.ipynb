```python
import rasterio
import matplotlib.pyplot as plt
```

```python
# This notebook explores a single 3 band RGB Sentinel-2 scene in a UTM projection.
image_file = "image1.tif"

# Use Rasterio to open the image.
satdat = rasterio.open(image_file)
```

```python
# Load the 3 bands into 2d arrays 
r, g, b = satdat.read()
```

```python
# Use imshow to load the blue band.
fig = plt.imshow(b)

# Display the results.
plt.show()
```
